# 3-Exploratory Data Analysis

# Question 1:The effect of the elements on the attendance or non-attendance of the patient

```python
def showdata(df,x,y) :
    print(df.groupby(x)[y].mean())
    df.groupby(x)[y].mean().plot(kind='bar');
    plt.xlabel("Showed")
    plt.ylabel(x)
    plt.title('relation between '+ x + ' and patients who showed')
    plt.legend()
```

##### 1-Scholarship

```python
showdata(df,'Scholarship','attend')
```

The patient who has a scholarship more not attended the reservation

#### 2-Hipertension

```python
showdata(df,'Hipertension','attend')
```

The patient who has a Hipertension more  attended the reservation

##### 3-Diabetes

```python
showdata(df,'SMS_received','attend')
```

The patient who has a Diabetes more not attended the reservation

##### 3-Gender

```python
showdata(df,'Gender','attend')
```

Gender did not affect attendance

##### 4-Alcoholism

```python
showdata(df,'Alcoholism','attend')
```

Alcoholism did not affect attendance
